```python
import maboss
```

Importing MaBoSS

### Setting the default simulation

```python
m = maboss.loadBNet("net_shah.bnet")
```

Loading the BNet file

```python
for node in m.network.keys():
    m.network.set_istate(node, [1, 0])
m.network.set_istate("LCK", [0, 1])
```

Setting up initial state. All at 0 except LCK at 1.

```python
for name, node in m.network.items():
    if name not in ["TAex", "IL2ex", "PDL1L2ex", "CD8086ex"]:
        node.set_rate(str(0.5), str(0.05))
```

Setting up rates : activation = 0.5, inactivation = 0.05

```python
m.network.set_output(["CFUNC", "CINHI", "TAPOP"])
```

Configuring output nodes : only showing CFUNC, CINHI, TAPOP in the results

```python
m.update_parameters(max_time=1000, sample_count=10000, thread_count=12)
```

Settings of the simulation: max_time = 1000, nb iterations = 10k, in parallel on 12 cores

### Scenario 1 

```python
m_1 = m.copy()
m_1.network.set_istate(["TAex", "IL2ex", "PDL1L2ex", "CD8086ex"], {(1,0,0,0): 1.0})
r_1 = m_1.run()
r_1.plot_node_trajectory()
```
